```python
text = '1000000_2_5'

file = 'CLEAN_' + text + '.txt'
path = 'data/' + text + '/'

with open(path + file, encoding="utf8") as handle:
    sentences =  [l.strip()[0:-1] for l in handle.readlines()]
    
    

sentences1 = []
for sent in sentences:
    sent = re.sub(r"\s+", " ", sent)
    sentences1.append(sent.strip())

x = []

for i in sentences1:
    if len(i.split()) > 2 and len(i.split()) < 50:
        x.append(i)
        
        
words= []
for h in sentences1:
    for i in h.split():
        words.append(len(i))

plt.xlabel("Word lenghts")
plt.ylabel("Amount of words")
plt.ylim(0, 200000)
plt.xlim(0, 30)
print(plt.hist(words, bins=35)[0])
```

```python
perplexities = []

for direc in os.listdir('data'):
    if direc[0].isnumeric():
        temp = []
        for direc1 in os.listdir('data/' + direc):
            if direc1 == '[model]':
                os.rename('data/' + direc + '/[model]', 'data/' + direc+"/model")
            if direc1[0].isnumeric():
                temp.append(int(direc1[0:3]))
            txtfile = codecs.open('data/' + direc + '/textinfo.txt','r',encoding='utf-8',errors='ignore')
            XD = txtfile.readlines()
            if XD[0][0] == 'V':
                parameter = int(XD[0][13:-1])
            if XD[1][0] == 'V':
                parameter = int(XD[1][13:-1])
        perplexities.append((parameter,temp))
        
perplexities = sorted(perplexities, key=lambda x: x[0])

x=[]
y=[]

for tup in perplexities:
    x.append(tup[0])
    y.append(tup[1][0])
    

plt.ylabel("Perplexity ")
plt.xlabel("Voc size")
plt.scatter(x,y, marker= 'x', color='red')     
```
